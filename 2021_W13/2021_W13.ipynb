```python
import pandas as pd
```

```python
# Input all the file
df1 = pd.read_csv("pl_15-16.csv")
df2 = pd.read_csv("pl_16-17.csv")
df3 = pd.read_csv("pl_17-18.csv")
df4 = pd.read_csv("pl_18-19.csv")
df5 = pd.read_csv("pl_19-20.csv")
```

```python
df = pd.concat([df1,df2,df3,df4,df5], axis=0)
df
```

```python
# Remove all goalkeepers from the data set
# Remove all records where appearances = 0	
# In this challenge we are interested in the goals scored from open play
# Create a new “Open Play Goals” field (the goals scored from open play is the number of goals scored that weren’t penalties or freekicks)
# Note some players will have scored free kicks or penalties with their left or right foot
# Be careful how Prep handles null fields! (have a look at those penalty and free kick fields) 
# Rename the original Goals scored field to Total Goals Scored


df = df[~(df['Position'] == 'Goalkeeper')]
df = df[~(df['Appearances'] == 0)]
df['Penalties scored'] = df['Penalties scored'].fillna(0)
df['Freekicks scored'] = df['Freekicks scored'].fillna(0)
df['Open Play Goals'] = df['Goals'] - df['Penalties scored'] - df['Freekicks scored']
df.rename(columns={'Goals':'Total Goals'}, inplace=True)
df
```

```python
df.columns
```

```python
# Calculate the totals for each of the key metrics across the whole time period for each player, (be careful not to lose their position)
# Create an open play goals per appearance field across the whole time period

df = df.groupby(['Name', 'Position']).agg({'Open Play Goals':'sum', 'Appearances':'sum', 'Total Goals':'sum', 'Headed goals':'sum', 'Goals with right foot':'sum', 'Goals with left foot':'sum'}).reset_index()

df
```

```python
# Rank the players for the amount of open play goals scored 
# across the whole time period, we are only interested in the top 20 
# (including those that are tied for position) – Output 1

output1 = df
output1['Open Play Goals per Appearance'] = output1['Open Play Goals'] / output1['Appearances']
output1['Rank'] = output1['Open Play Goals'].rank(method='min', ascending=False)
output1 = output1[output1['Rank'] <= 20]

output1
```

```python
# Rank the players for the amount of open play goals scored across the whole time 
# period by position, we are only interested in the top 20 (including those that 
# are tied for position) – Output 2
# Output the data – in your solution on twitter / the forums, state the name of the player 
# who was the only non-forward to make it into the overall top 20 for open play 
# goals scored

output2 = df
output2['Rank'] = output2.groupby('Position')['Open Play Goals'].rank(method='min', ascending=False)
output2 = output2[output2['Rank'] <= 20]
output2
```
