```python
import pandas as pd

# Read the CSV files
df1 = pd.read_csv('Prep Air Ticket Sales.csv')
df2 = pd.read_csv('Prep Air 2024 Flights.csv')
df3 = pd.read_csv('Prep Air Customers.csv')

# Display the first few rows of each dataframe to verify
print(df1.head())
print(df2.head())
print(df3.head())
```

```python
df_merged = pd.merge(df1, df2, on=['Date', 'Flight Number'], how='left')
print(df_merged.head())
```

```python
output1 = pd.merge(df_merged, df3, on='Customer ID', how='left')
print(output1)
```

```python
# Identify unbooked flights
unbooked_flights = df2[~df2[['Date', 'Flight Number']].apply(tuple, axis=1).isin(df1[['Date', 'Flight Number']].apply(tuple, axis=1))]

# Add a datestamp field for today's date (31/01/2024)
unbooked_flights['Flight unbooked as of'] = '2024-01-31'

print(unbooked_flights)
```

```python
# Merge df3 and df1 to find customers who haven't booked any flight in 2024
customers_no_flight_2024 = pd.merge(df3, df1, on='Customer ID', how='left', indicator=True)

# Filter out customers who have booked flights
customers_no_flight_2024 = customers_no_flight_2024[customers_no_flight_2024['_merge'] == 'left_only']

# Drop unnecessary columns
customers_no_flight_2024 = customers_no_flight_2024.drop(columns=['Date', 'Flight Number', 'Ticket Price', '_merge'])

print(customers_no_flight_2024)
```

```python
# Convert 'Last Date Flown' to datetime
customers_no_flight_2024['Last Date Flown'] = pd.to_datetime(customers_no_flight_2024['Last Date Flown'])

# Reference date
reference_date = pd.to_datetime('2024-01-31')

# Calculate days since last flown
customers_no_flight_2024['Days Since Last Flown'] = (reference_date - customers_no_flight_2024['Last Date Flown']).dt.days

print(customers_no_flight_2024)
```

```python
# Categorize customers based on days since last flown
def categorize_customer(days):
    if days <= 90:
        return 'Recent fliers'
    elif 90 < days <= 180:
        return 'Taking a break'
    elif 180 < days <= 270:
        return 'Been away a while'
    else:
        return 'Lapsed Customers'

customers_no_flight_2024['Customer Category'] = customers_no_flight_2024['Days Since Last Flown'].apply(categorize_customer)

print(customers_no_flight_2024)
```

```python
output3 = customers_no_flight_2024
output2 = unbooked_flights
output1 = output1
```
