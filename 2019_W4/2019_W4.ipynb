```python
import pandas as pd
```

```python
source = 'PD - ESPN stats.xlsx'
converter= {}
df = pd.read_excel(source, sheet_name='Sheet1')
df.head(10)
```

Split column

```python
df[['HI ASSISTS - Player','HI ASSISTS - Value']] = df['HI ASSISTS'].str.split(' ', expand=True)
df[['HI REBOUNDS - Player','HI REBOUNDS - Value']] = df['HI REBOUNDS'].str.split(' ', expand=True)
df[['HI POINTS - Player','HI POINTS - Value']] = df['HI POINTS'].str.split(' ', expand=True)
df.head()
```

```python
# df['OPPONENT'] = df['OPPONENT'].str.replace(r'^[v][s].*', '')
# df.head()
```

Create 'Home pr Away' , 'Opponent (clean)' , 'Win or Loss' columns

```python
home_or_away = []
cleaned_opponent = []
win_or_lose = []
for index, col in df.iterrows():
    if df.loc[index,'OPPONENT'].startswith('vs'):
        home_or_away.append('Home')
        cleaned_opponent.append(df.loc[index,'OPPONENT'][2:])
    else:
        home_or_away.append('Away')
        cleaned_opponent.append(df.loc[index,'OPPONENT'][1:])

    if df.loc[index, 'RESULT'].startswith('W'):
        win_or_lose.append('W')
    else:
        win_or_lose.append('L')
df['Home or Away'] = home_or_away
df['Opponent (clean)'] = cleaned_opponent
df['Win or Loss'] = win_or_lose
df.head()
```

Remove unnessesary columns, rename 'DATE' into 'True Date'

```python
df.drop(['HI POINTS', 'HI REBOUNDS','HI ASSISTS'], axis=1, inplace=True)
df.rename(columns={'DATE':'True Date'}, inplace=True)
df.head()
```

```python
with pd.ExcelWriter('Output.xlsx', mode='w') as write:
    df.to_excel(write, index=False)
```
