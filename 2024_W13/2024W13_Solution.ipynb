```python
# Merge df_all with date_df on 'Sales Date' to get 'Week Number'
df_all = df_all.merge(date_df, left_on='Sales Date', right_on='Date', how='left')

# Drop the redundant 'Date' column from the merged dataframe
df_all = df_all.drop(columns=['Date'])

df_all
```

```python
# Create a new column 'Year' from 'Sales Date'
df_all['Year'] = df_all['Sales Date'].dt.year

df_all
```

```python
# Import necessary library
import pandas as pd

# Create the 'Easter Week Number' column
df_all['Easter Week Number'] = df_all.groupby('Year')['Week Number'].rank(method='dense').astype(int)

df_all
```

```python
# Add a new column to indicate the day of the week for each Sales Date
df_all['Weekday'] = df_all['Sales Date'].dt.day_name()

df_all
```

```python
df_all['Weekday'] = df_all['Weekday'].apply(lambda x: x[:2] if x.startswith(('T', 'S')) else x[0])
df_all
```

```python
# Rank the sales dates in ascending order, partitioned by year and Easter week number
df_all['Weekday Order'] = df_all.groupby(['Year', 'Easter Week Number'])['Sales Date'].rank(method='dense').astype(int)

df_all
```

```python
# Create a new column 'Sale' which is Quantity Sold * Price
df_all['Sale'] = df_all['Quantity Sold'] * df_all['Price']

df_all
```
