The aim of this challenge is to group Pokémon into Evolution Groups. Why? So that when we select any Pokémon, we can see all the information about what they evolved from, what they will evolve to etc. In the example below, Bulbasaur evolves into Ivysaur, who then evolves into Venusaur. How is this applicable to other hierarchical data? Well imagine that a Pokémon that has a null in the Evolves From field is similar to an employee at Work Level 1 in an organisation. Their manager would then sit in the Evolves To field. Our Evolution Group in this instance would be the Team the employee belongs to and it would give us the ability to type in any employee name and return information about which team they belong to and the structure of that team.

```python
import pandas as pd
import numpy as np
```

```python
input = 'Pokemon Input.xlsx'
sheets = pd.ExcelFile(input).sheet_names
sheets
```

```python
pokemon = pd.read_excel(input, 'Pokemon')
# Correct data type
pokemon['#'] = pokemon['#'].astype(float)
print(pokemon.head(5))
print(pokemon.info())
print(pokemon.isnull().any())
```

```python
evolution = pd.read_excel(input, 'Evolution')
print(evolution.head(5))
print(evolution.info())
print(evolution.isnull().any())
```

```python
# Our Pokémon dataset actually contains too many Pokémon:
# We're only interested in Pokémon up to Generation III, which is up to (and including) number 386
# This means we're also not interested in mega evolutions so we can filter Pokémon whose name start with "Mega"

pokemon = pokemon[pokemon['#'] <= 386]
pokemon = pokemon[~pokemon['Name'].str.startswith('Mega ')]
pokemon
```

```python
# Some Pokémon have more than one Type. 
# We aren't interested in Types for this challenge so remove this field 
# and ensure we have one row per Pokémon
pokemon.drop(columns='Type', inplace=True)
pokemon.drop_duplicates(inplace=True)
pokemon
```

```python
# Now we want to bring in information about what our Pokémon evolve to
# Warning!  In our Evolution dataset, we still have Pokémon beyond Gen III. 
# You'll need to filter these out too, from both the evolved from and evolved to fields
pokemon_gen_123_list = pokemon['Name'].tolist()
evolution = evolution[evolution['Evolving to'].isin(pokemon_gen_123_list)]
evolution = evolution[evolution['Evolving from'].isin(pokemon_gen_123_list)]
evolution
```

```python
# Bring in information about what a Pokémon evolves from
# Ensure that we have all 386 of our Pokémon, with nulls if they don't 
# have a pre-evolved form or if they don't evolve

detail_df = pd.merge(left=pokemon, right=evolution, left_on='Name', right_on='Evolving from', how='left')
detail_df.drop(columns='Evolving from', inplace=True)
detail_df = pd.merge(left=detail_df, right=evolution[['Evolving from','Evolving to']], left_on='Name', right_on='Evolving to', how='left')
detail_df.drop(columns='Evolving to_y', inplace=True)
detail_df.rename(columns={'Evolving to_x':'Evolving to'}, inplace=True)
detail_df.drop_duplicates(inplace=True)
detail_df
# detail_df[detail_df['Name'] == 'Eevee']
```

```python
# Create a calculation for our Evolution Group
# The Evolution Group will be named after the First Evolution e.g. in the above example,
# Bulbasaur is the name of the Evolution Group



detail_df['First Evolution'] = detail_df.apply(lambda row: row['Name'] if pd.isnull(row['Evolving from']) 
                                               else np.nan, axis=1)
evolution_tree = detail_df['First Evolution'].dropna().drop_duplicates()
evolution_tree = pd.merge(left=evolution_tree, right=evolution[['Evolving from','Evolving to']], left_on='First Evolution', right_on='Evolving from', how='left')
evolution_tree.drop(columns='Evolving from', inplace=True)
evolution_tree.rename(columns={'Evolving to':'Second Evolution'}, inplace=True)
evolution_tree = pd.merge(left=evolution_tree, right=evolution[['Evolving from','Evolving to']], left_on='Second Evolution', right_on='Evolving from', how='left')
evolution_tree.drop(columns='Evolving from', inplace=True)
evolution_tree.rename(columns={'Evolving to':'Third Evolution'}, inplace=True)
evolution_tree
```

```python
evolution_tree.loc[0].tolist()
```

```python
for index, row in detail_df.iterrows():
    if pd.isnull(row['Evolving from']):
        detail_df.at[index, 'First Evolution'] = row['Name']
    else:
        for i, r in evolution_tree.iterrows():
            if row['Evolving from'] in evolution_tree.loc[i].tolist():
                detail_df.at[index, 'First Evolution'] = evolution_tree.loc[i].tolist()[0]
        
detail_df.columns
```

```python
output = detail_df[['First Evolution', '#', 'Name', 'Total', 'HP', 'Attack', 'Defense', 'Special Attack',
       'Special Defense', 'Speed', 'Evolving from', 'Evolving to', 'Level', 'Condition',
       'Evolution Type']]
output
```
