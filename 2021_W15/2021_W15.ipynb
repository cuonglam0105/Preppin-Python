```python
import pandas as pd
```

```python
input = 'Menu and Orders.xlsx'
excel_sheets = pd.ExcelFile(input).sheet_names
excel_sheets
```

```python
menu = pd.read_excel(input, sheet_name='MENU')
menu
```

```python
order = pd.read_excel(input, sheet_name='Order')
order
```

```python
menu.columns = menu.columns.str.strip()
menu.columns
```

```python
# Modify the structure of the Menu table so we can have one column for the Type (pizza, pasta, house plate), the name of the plate, ID, and Price (hint)
menu1 = menu[['Pizza', 'Pizza Price', 'Pizza ID']].copy()
menu1.rename(columns={'Pizza':'Name', 'Pizza Price':'Price', 'Pizza ID':'ID'}, inplace=True)
menu1['Type'] = 'Pizza'
menu2 = menu[['Pasta', 'Pasta Price', 'Pasta ID']].copy()
menu2.rename(columns={'Pasta':'Name', 'Pasta Price':'Price', 'Pasta ID':'ID'}, inplace=True)
menu2['Type'] = 'Pasta'
menu3 = menu[['House Plates', 'House Plates Prices', 'House Plates ID']].copy()
menu3.rename(columns={'House Plates':'Name', 'House Plates Prices':'Price', 'House Plates ID':'ID'}, inplace=True)
menu3['Type'] = 'House Plates'

menu_df = pd.concat([menu1,menu2,menu3], axis=0, ignore_index=True)
menu_df.dropna(how='any', inplace=True)
menu_df['ID'] = menu_df['ID'].astype(int)
menu_df
```

```python
# Modify the structure of the Orders table to have each item ID in a different row (hint)
order[['0','1','2','3']] = pd.DataFrame(order['Order'].str.split('-',expand=True))
order['0'].fillna(order['Order'], inplace=True)
order.drop(columns='Order', inplace=True)

order = order.melt(id_vars=['Order Date', 'Customer Name'], value_vars=['0','1','2','3'], var_name='pivot_name', value_name='Item ID')
order.dropna(inplace=True)
order.drop(columns='pivot_name', inplace=True)
order['Item ID'] = order['Item ID'].astype(int)
order
```

```python
# Join both tables (hint)
join_df = pd.merge(left=order, right=menu_df, how='inner', left_on='Item ID', right_on='ID')
join_df
```

```python
# On Monday's we offer a 50% discount on all items. Recalculate the prices to reflect this
join_df['Day of week'] = join_df['Order Date'].dt.day_name()
join_df['Price'] = join_df.apply(lambda row: row['Price'] / 2 if row['Day of week'] == 'Monday' else row['Price'], axis=1)
join_df['Price'] = join_df['Price'].astype(int)

join_df
```

```python
# For Output 1, we want to calculate the total money for each day of the week (hint)
output1 = join_df.groupby('Day of week').agg(Total_price=('Price','sum')).reset_index()
output1
```

```python
# For Output 2, we want to reward the customer who has made the most orders for their loyalty.
# Work out which customer has ordered the most single items. (hint)
output2 = join_df.groupby(['Customer Name']).agg(count_item=('Item ID','count')).reset_index()
output2 = output2[output2['count_item'] == output2['count_item'].max()]

output2
```
