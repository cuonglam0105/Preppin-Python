```python
import pandas as pd

# Read the Excel file
excel_file = 'store stock data advanced.xlsx'

# List all sheets
sheets = pd.ExcelFile(excel_file).sheet_names
print(sheets)
```

```python
ad_stock_df = pd.read_excel(excel_file, sheet_name='Adanced Stock per Store Data')
ad_stock_df
```

```python
recall_df = pd.read_excel(excel_file, sheet_name='Recall Items')
recall_df.head()
```

```python
merged_df = pd.merge(ad_stock_df, recall_df, on=['Category', 'Product ID'], how='inner')
merged_df.head()
```

```python
# Drop the 'Unit Price_y' column
merged_df.drop(columns=['Unit Price_y'], inplace=True)

# Rename 'Unit Price_x' to 'Unit Price'
merged_df.rename(columns={'Unit Price_x': 'Unit Price'}, inplace=True)

# Display the updated dataframe
merged_df
```

```python
# Define the recall announcement date
recall_date = pd.to_datetime('2025-05-13')

# Calculate the number of days taken to remove the items
merged_df['Days to Remove'] = (merged_df['Date'] - recall_date).dt.days

# Display the updated dataframe
merged_df
```

```python
# Define the target recall date
target_date = pd.to_datetime('2025-05-20')

# Categorize the items
def categorize_recall(row):
    if pd.isna(row['Days to Remove']):
        return 'Incomplete'
    elif row['Days to Remove'] <= (target_date - recall_date).days:
        return 'On Target'
    else:
        return 'Overdue'

merged_df['Recall Status'] = merged_df.apply(categorize_recall, axis=1)

# Display the updated dataframe
merged_df
```

```python
# Group by store and calculate the average days to remove
average_days_to_remove = merged_df.groupby('Store')['Days to Remove'].mean().reset_index()

# Rename the column for clarity
average_days_to_remove.rename(columns={'Days to Remove': 'Average Days to Remove'}, inplace=True)

# Display the result
average_days_to_remove
```

```python
# Convert the average days to remove into separate columns for days and hours
average_days_to_remove['Days to Removal'] = average_days_to_remove['Average Days to Remove'].astype(int)
average_days_to_remove['Hours to Removal'] = ((average_days_to_remove['Average Days to Remove'] - average_days_to_remove['Days to Removal']) * 24).round().astype(int)

# Drop the 'Average Days to Remove' column as it's no longer needed
average_days_to_remove.drop(columns=['Average Days to Remove'], inplace=True)

# Display the updated dataframe
average_days_to_remove
```

```python
# Calculate the number of days overdue
filtered_df['Days Overdue'] = filtered_df.apply(lambda row: row['Days to Remove'] - (target_date - recall_date).days if row['Recall Status'] == 'Overdue' else None, axis=1)

# Display the updated dataframe
filtered_df.head()
```
