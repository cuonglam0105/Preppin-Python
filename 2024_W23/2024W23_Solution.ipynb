```python
import pandas as pd

# Read the Excel file
excel_file = pd.ExcelFile('Superbyte Managers (Beginner).xlsx')

# List the sheet names
sheet_names = excel_file.sheet_names
print(sheet_names)
```

```python
tenures_df = pd.read_excel(excel_file, sheet_name='Tenures')
tenures_df.head()
```

```python
salaries_df = pd.read_excel(excel_file, sheet_name='Salaries')
salaries_df.head()
```

```python
amount_df = pd.read_excel(excel_file, sheet_name='Amount Paid')
amount_df.head()
```

```python
# title() capitalizes the first letter of each word
salaries_df['Name'] = salaries_df['Name'].str.title()
salaries_df.head()
```

```python
# title() capitalizes the first letter of each word
tenures_df['Name'] = tenures_df['Name'].str.title()
tenures_df.head()
```

```python
# Merge the two DataFrames
merged_df = tenures_df.merge(salaries_df, on='Name', how='left')
merged_df.head()
```

```python
# Merge the two DataFrames
merged_df['End Date'] = merged_df['End Date'].replace('Current', '2024-06-05')
merged_df.head()
```

```python
# Convert the 'End Date' column to datetime
merged_df['End Date'] = pd.to_datetime(merged_df['End Date'])
merged_df.head()
```

```python
# Calculate the number of days worked
merged_df['Days Worked'] = (merged_df['End Date'] - merged_df['Start Date']).dt.days
merged_df.head()
```

```python
# Merge the two DataFrames
merged_df['Expected Total Salary'] = (merged_df['Salary'] / 365) * merged_df['Days Worked']
merged_df['Expected Total Salary'] = merged_df['Expected Total Salary'].round()
merged_df.head()
```

```python
# Merge the two DataFrames
final_df = merged_df.merge(amount_df, on='Name', how='left')
final_df.head()
```

```python
# Check for salary mismatches
final_df['Salary Mismatch'] = final_df['Expected Total Salary'] != final_df['Amount Paid']
final_df.head()
```

```python
# Filter the mismatched salaries
mismatched_salaries_df = final_df[final_df['Salary Mismatch'] == True]
output = mismatched_salaries_df.drop(columns=['Salary Mismatch'])
output
```

```python
# filter to only show the mismatched salaries which have an end date of 2024-06-05
output = output[output['End Date'] == '2024-06-05']
output
```
