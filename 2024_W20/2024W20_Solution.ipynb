```python
import pandas as pd

# Read the Excel file
file_path = 'SuperBytes Customer Data.xlsx'
excel_data = pd.ExcelFile(file_path)

# List the sheet names
sheet_names = excel_data.sheet_names
print(sheet_names)
```

```python
# Read the Customer Spending sheet with specific header rows
customer_spending_df = pd.read_excel(file_path, sheet_name='Customer Spending', header=[0, 1])

# Combine the first and second column headers
customer_spending_df.columns = [' '.join(col).strip() for col in customer_spending_df.columns.values]

customer_spending_df
```

```python
# Remove 'Last ' from column names except for the first two columns
customer_spending_df.columns = [col if i < 2 else col.replace('Last ', '') for i, col in enumerate(customer_spending_df.columns)]

customer_spending_df
```

```python
# Read the Customer Names sheet
customer_df = pd.read_excel(file_path, sheet_name='Customer Names')

customer_df
```

```python
# Merge the dataframes on the common column (assuming 'Customer ID' is the common column)
merged_df = pd.merge(customer_spending_df, customer_df, on=['First Name', 'Last Name'], how='inner')

merged_df
```

```python
# Drop 'First Name' and 'Last Name' columns
merged_df = merged_df.drop(columns=['First Name', 'Last Name'])

merged_df
```

```python
# Create a new field 'Online or In Person' based on 'Online' and 'In Person' columns
merged_df['Online or In Person'] = merged_df.apply(lambda row: 'Online' if row['Online'] == 'Yes' else 'In Person', axis=1)

# Drop 'Online' and 'In Person' columns
merged_df = merged_df.drop(columns=['Online', 'In Person'])

merged_df
```

```python
# Add a new column 'Weekday' based on the 'Date' column
merged_df['Weekday'] = merged_df['Date'].dt.day_name()

merged_df
```

```python
# Rank sales for each weekday in descending order
merged_df['Rank'] = merged_df.groupby('Weekday')['Sale Total'].rank(ascending=False, method='dense')

# Sort dataset by weekday
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
merged_df['Weekday'] = pd.Categorical(merged_df['Weekday'], categories=weekday_order, ordered=True)

# Sort dataset by weekday
merged_df = merged_df.sort_values(by=['Weekday', 'Rank'])

# Drop 'Date' column
merged_df = merged_df.drop(columns=['Date'])

# Rearrange columns
merged_df = merged_df[['Rank', 'Customer ID', 'Receipt Number', 'Gender', 'Online or In Person', 'Sale Total', 'Weekday']]

output = merged_df
output
```
