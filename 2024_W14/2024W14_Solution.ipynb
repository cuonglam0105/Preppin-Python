```python
import pandas as pd

# Read the Excel file
excel_file = 'store stock beginner data.xlsx'

# List all sheet names
sheet_names = pd.ExcelFile(excel_file).sheet_names
print(sheet_names)
```

```python
stock_df = pd.read_excel(excel_file, sheet_name='Stock per Store')
stock_df.head(5)
```

```python
recalled_df = pd.read_excel(excel_file, sheet_name='Recalled Items')
recalled_df.head(5)
```

```python
# Merge the two dataframes
merged_df = pd.merge(stock_df, recalled_df, on=['Product ID', 'Category'], how='inner')
merged_df
```

```python
# drop the 'Unit Price_y' column and rename 'Unit Price_x' to 'Unit Price'
merged_df = merged_df.drop(columns=['Unit Price_y']).rename(columns={'Unit Price_x': 'Unit Price'})
merged_df.head()
```

```python
merged_df['Total loss'] = merged_df['Unit Price'] * merged_df['Quantity']
merged_df.head()
```

```python
category_loss = merged_df.groupby('Category')['Total loss'].sum().round(2).reset_index()
category_loss = category_loss.rename(columns={'Total loss': 'Total Price Rounded'})
output1 = category_loss
```

```python
store_loss = merged_df.groupby('Store')['Total loss'].sum().round(2).reset_index()
store_loss = store_loss.rename(columns={'Total loss': 'Total Price Rounded'})
store_loss
```

```python
store_loss['Issue Level'] = store_loss['Total Price Rounded'].apply(lambda x: 'High Priority' if x >= 5000 else 'Low Priority')
output2 = store_loss
output2
```
