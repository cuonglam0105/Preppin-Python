```python
climate_df
```

```python
climate_df = climate_df[['High °F', 'Low °F']]
climate_df
```

```python
# Aggregate data to get the minimum and maximum temperatures in Celsius
san_diego_climate = climate_df.agg({'High °F': 'max', 'Low °F': 'min'})
san_diego_climate
```

```python
# split the habitat column and explode the dataframe
output1['Habitat-1'] = output1['Habitat']
output1['Habitat'] = output1['Habitat'].str.split(',')
output1 = output1.explode('Habitat').reset_index(drop=True)
output1['Habitat'] = output1['Habitat'].str.strip()
output1
```

```python
habitat_df = habitat_df[['Habitat', 'Min Temp', 'Max Temp']]

habitat_df
```

```python
habitat_df['Min Temp'] = habitat_df['Min Temp'].str.replace('–', '-')
habitat_df['Max Temp'] = habitat_df['Max Temp'].str.replace('–', '-')
habitat_df
```

```python
habitat_df['Max Temp'] = pd.to_numeric(habitat_df['Max Temp'].str.extract(r'(-?\d+)')[0])
habitat_df['Min Temp'] = pd.to_numeric(habitat_df['Min Temp'].str.extract(r'(-?\d+)')[0])
habitat_df
```

```python
habitat_df.rename(columns={'Min Temp': 'Min Temp °F', 'Max Temp': 'Max Temp °F'}, inplace=True)
habitat_df
```

```python
merged_habitat_df = pd.merge(output1, habitat_df, on='Habitat', how='left', indicator=True)
merged_habitat_df
```

```python
# Group by Wildlife and Name, and calculate the min and max temperatures
wildlife_temp_df = merged_habitat_df.groupby(['Wildlife', 'Name', 'Habitat-1','Region']).agg(
    Min_Temp_Wildlife=('Min Temp °F', 'min'),
    Max_Temp_Wildlife=('Max Temp °F', 'max')
).reset_index()

wildlife_temp_df
```

```python
# Add San Diego climate data to wildlife_temp_df
wildlife_temp_df['San_Diego_Min_Temp'] = san_diego_climate['Low °F']
wildlife_temp_df['San_Diego_Max_Temp'] = san_diego_climate['High °F']

wildlife_temp_df
```

```python
def calculate_degrees_outside_ideal(row):
    if row['San_Diego_Min_Temp'] < row['Min_Temp_Wildlife']:
        return row['San_Diego_Min_Temp'] - row['Min_Temp_Wildlife']
    elif row['San_Diego_Max_Temp'] > row['Max_Temp_Wildlife']:
        return row['San_Diego_Max_Temp'] - row['Max_Temp_Wildlife']
    else:
        return 0

wildlife_temp_df['Degrees_outside_Ideal'] = wildlife_temp_df.apply(calculate_degrees_outside_ideal, axis=1)
wildlife_temp_df
```

```python
wildlife_temp_df['Habitat_Notes'] = wildlife_temp_df['Degrees_outside_Ideal'].apply(
    lambda x: 'Ideal' if x == 0 else ('Below' if x < 0 else 'Above')
)
wildlife_temp_df
```

```python
non_ideal_habitat_df = wildlife_temp_df[wildlife_temp_df['Habitat_Notes'] != 'Ideal']
non_ideal_habitat_df
```

```python
priority_df
```

```python
unique_degrees_outside_ideal = non_ideal_habitat_df['Degrees_outside_Ideal'].unique()
print(unique_degrees_outside_ideal)
```

```python
non_ideal_habitat_df['Absolute_Degrees_outside_Ideal'] = non_ideal_habitat_df['Degrees_outside_Ideal'].abs()
non_ideal_habitat_df['Priority_Order'] = non_ideal_habitat_df['Absolute_Degrees_outside_Ideal'].rank(method='dense', ascending=False)
non_ideal_habitat_df
```

```python
priority_df
```

```python
# Merge non_ideal_habitat_df with priority_df on the 'Status' column
final_df = pd.merge(non_ideal_habitat_df, priority_df, left_on='Priority_Order', right_on='Priority', how='left')

# If a wildlife species doesn't have a Status, make their Priority number 6
final_df['Priority'] = final_df['Priority'].fillna(6)

final_df
```

```python
final_df_filtered = final_df.drop(columns=['Min_Temp_Wildlife', 'Max_Temp_Wildlife', 'San_Diego_Min_Temp', 'San_Diego_Max_Temp', 'Degrees_outside_Ideal', 'Priority'])
final_df_filtered
```

```python
final_df_filtered = final_df_filtered.merge(merged_df[['Name', 'Class']], on='Name', how='left')
output2 = final_df_filtered
output2
```
